### Setup (constants + helpers)

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
with mpl.rc_context({"text.usetex": False}): 
    fig, ax = plt.subplots()

import matplotlib.pyplot as plt

from pathlib import Path

from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------
# Core constants
# -----------------------------
DT_SECONDS   = 4          # your logging cadence (4 s)
LOOKBACK_MIN = 10         # past window
WIN = int((LOOKBACK_MIN * 60) / DT_SECONDS)   # 150 samples

HORIZON_S = 60            # fixed as you decided
K = int(HORIZON_S / DT_SECONDS)               # 15 steps

PHASES = ["A", "B", "C"]

TRIPLEN = [3, 9, 15]
ODD     = [5, 7, 11]
ORDERS  = sorted(list(set(TRIPLEN + ODD)))    # [3,5,7,9,11,15]

# Detection thresholds (paper)
THD_WATCH_LO = 6.4
THD_EXCEED   = 8.0

# Current “behavior” KPI thresholds (NOT IEEE compliance unless you have TDD, Isc/IL)
IHD_WATCH_LO = 4.0
IHD_EXCEED   = 5.0

def paper_axes(ax, title=None, xlabel="Time", ylabel=None):
    """Conference-friendly axes formatting."""
    ax.grid(True, alpha=0.3)
    ax.set_xlabel(xlabel)
    if ylabel: ax.set_ylabel(ylabel)
    if title: ax.set_title(title)
    for spine in ["top", "right"]:
        ax.spines[spine].set_visible(False)

def legend_outside(ax, ncol=1):
    """Place legend outside right (paper-friendly, avoids covering curves)."""
    ax.legend(loc="upper left", bbox_to_anchor=(1.02, 1.0), borderaxespad=0, ncol=ncol)

def safe_numeric(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

def max_over_phases(df, colsA, colsB, colsC):
    """Convenience to take max across phase columns that represent same signal per phase."""
    existing = [c for c in (colsA + colsB + colsC) if c in df.columns]
    return df[existing].max(axis=1) if existing else pd.Series(np.nan, index=df.index)
```

### Load voltage & current CSVs + make timestamps

```python
# Prefer augmented files if present; else fall back to originals
df_v = pd.read_csv("VoltageHarmonics_1_aug20.csv", skiprows=2)
df_i = pd.read_csv("CurrentHarmonics_1_aug20.csv", skiprows=2)

# Timestamp creation - explicitly specify the format to match your data
df_v["Timestamp"] = pd.to_datetime(df_v["Date"].astype(str) + " " + df_v["Time"].astype(str),
                                   format="%d/%m/%Y %H:%M:%S", errors="coerce")
df_i["Timestamp"] = pd.to_datetime(df_i["Date"].astype(str) + " " + df_i["Time"].astype(str),
                                   format="%m/%d/%Y %H:%M:%S", errors="coerce")

df_v = df_v.dropna(subset=["Timestamp"]).sort_values("Timestamp").reset_index(drop=True)
df_i = df_i.dropna(subset=["Timestamp"]).sort_values("Timestamp").reset_index(drop=True)

print("Voltage:", df_v.shape, "| Current:", df_i.shape)
print("Span:", df_v["Timestamp"].min(), "→", df_v["Timestamp"].max())
print("Δt head counts:\n", df_v["Timestamp"].diff().value_counts().head())
```

### Merge V+I into one master table (time-aligned)

```python
df = pd.merge(df_v, df_i, on="Timestamp", how="inner", suffixes=("", "_Idup"))

# Remove accidental duplicates created by merge (same column names)
dup_cols = [c for c in df.columns if c.endswith("_Idup")]
df = df.drop(columns=dup_cols)

df = df.sort_values("Timestamp").drop_duplicates("Timestamp").set_index("Timestamp")

print("Merged df:", df.shape)
```

### Select columns (target harmonics) + numeric conversion

```python
# ==============================
# Cell 3 (UPDATED): Phase-A only
# ==============================
PH = "A"

# Voltage UHD (Phase A)
U_trip_cols_A = [f"UHD{h}{PH}" for h in [3, 9, 15]]
U_odd_cols_A  = [f"UHD{h}{PH}" for h in [5, 7, 11]]

# Current IHD (Phase A)
I_odd_cols_A  = [f"IHD{h}{PH}" for h in [5, 7, 11]]

# THD (Phase A)
uthd_cols = [f"UTHD{PH}"]   # ["UTHDA"]
ithd_cols = [f"ITHD{PH}"]   # ["ITHDA"]

# -------- Inputs to ML (past 10 min features built from these) --------
X_base_cols = U_trip_cols_A + U_odd_cols_A + I_odd_cols_A + uthd_cols + ithd_cols

# -------- Outputs from ML (targets at t+60s) --------
# This is the critical change: we ADD UHD and ITHD to the targets.
Y_cols = uthd_cols + U_trip_cols_A + U_odd_cols_A + ithd_cols + I_odd_cols_A

# Ensure numeric
safe_numeric(df, X_base_cols + Y_cols)

missing_inputs = [c for c in X_base_cols if c not in df.columns]
missing_outputs = [c for c in Y_cols if c not in df.columns]

print("Missing input cols:", missing_inputs)
print("Missing output cols:", missing_outputs)
print("Using inputs:", len([c for c in X_base_cols if c in df.columns]),
      "| Using outputs:", len([c for c in Y_cols if c in df.columns]))
print("Targets:", Y_cols)
```

# Detection (NOW): 2 plots (Triplen + Odd)

### Build NOW “behavior indices” for Triplen and Odd IHD

```python
def ihd_index(df_now, orders):
    cols = [f"IHD{h}{ph}" for h in orders for ph in PHASES if f"IHD{h}{ph}" in df_now.columns]
    return df_now[cols].max(axis=1) if cols else pd.Series(np.nan, index=df_now.index)

trip_now = ihd_index(df, TRIPLEN)
odd_now  = ihd_index(df, ODD)

# Also handy: current voltage health now (max UTHD across phases)
uthd_now = df[uthd_cols].max(axis=1) if all(c in df.columns for c in uthd_cols) else pd.Series(np.nan, index=df.index)

print("Triplen NOW index stats:", trip_now.describe()[["min","mean","max"]].to_dict())
print("Odd NOW index stats:", odd_now.describe()[["min","mean","max"]].to_dict())
```

### Auto-pick best window for the NOW detection plots

```python
# ============================================================
# AUTO-PICK BEST "NOW" WINDOW (for Triplen/Odd detection plots)
# ============================================================

WINDOW_MIN = 15  # choose 10–20 min; 15 min usually looks best in paper figures
W = int((WINDOW_MIN * 60) / DT_SECONDS)

def _state_score(series, watch_lo, exceed_hi, w_watch=1.0, w_exceed=3.0):
    """Convert a signal into a per-sample score based on watch/exceed thresholds."""
    s = series.astype(float)
    watch = (s >= watch_lo) & (s < exceed_hi)
    exceed = (s >= exceed_hi)
    return (w_watch * watch.astype(int) + w_exceed * exceed.astype(int))

def pick_best_window(score_series, window_samples):
    """Return start/end of the highest-scoring rolling window."""
    roll = score_series.rolling(window_samples, min_periods=window_samples).sum()
    if roll.dropna().empty:
        raise RuntimeError("Not enough samples to form the requested window.")
    end = roll.idxmax()
    start = end - pd.Timedelta(seconds=(window_samples - 1) * DT_SECONDS)
    return start, end, float(roll.loc[end])

# Build a combined "interestingness" score (NOW)
score_now = pd.Series(0.0, index=df.index)

# Current behavior (KPI thresholds)
score_now += _state_score(trip_now, IHD_WATCH_LO, IHD_EXCEED, w_watch=1.0, w_exceed=3.0)
score_now += _state_score(odd_now,  IHD_WATCH_LO, IHD_EXCEED, w_watch=1.0, w_exceed=3.0)

# Voltage health (standards thresholds: THD watch/exceed)
score_now += _state_score(uthd_now, THD_WATCH_LO, THD_EXCEED, w_watch=1.0, w_exceed=4.0)

BEST_START_NOW, BEST_END_NOW, BEST_SCORE_NOW = pick_best_window(score_now, W)
print(f"Best NOW window: {BEST_START_NOW} → {BEST_END_NOW} | score={BEST_SCORE_NOW:.1f} | window={WINDOW_MIN} min")

# Convenience index to use in plots
IDX_NOW = df.loc[BEST_START_NOW:BEST_END_NOW].index
```

# PART B — Prediction (ML) + Results Table

### Signal-processing style features (10-min window)

```python
base = df[X_base_cols].copy()

# Rolling window statistics (telemetry-first “signal processing”)
roll = base.rolling(window=WIN, min_periods=WIN)

X = pd.concat({
    "last": base,
    "mean10m": roll.mean(),
    "std10m": roll.std(),
    "min10m": roll.min(),
    "max10m": roll.max(),
    "trend10m": base - base.shift(WIN-1),   # last - first (10 min)
    "delta60s": base - base.shift(K),       # last - value 60s ago
}, axis=1)

# Flatten column MultiIndex
X.columns = [f"{grp}__{col}" for grp, col in X.columns]

print("Feature matrix X:", X.shape)
```

### Build targets at +60s, align dataset, time split

```python
Y = df[Y_cols].shift(-K).copy()
Y.columns = [f"{c}__t+{HORIZON_S}s" for c in Y.columns]

data = pd.concat([X, Y], axis=1).dropna()

X_all = data[X.columns]
Y_all = data[Y.columns]

# Blocked split: 70/15/15 (time-ordered, no leakage)
n = len(data)
n_train = int(0.70 * n)
n_val   = int(0.15 * n)

X_train, Y_train = X_all.iloc[:n_train], Y_all.iloc[:n_train]
X_val,   Y_val   = X_all.iloc[n_train:n_train+n_val], Y_all.iloc[n_train:n_train+n_val]
X_test,  Y_test  = X_all.iloc[n_train+n_val:], Y_all.iloc[n_train+n_val:]

print("Train/Val/Test:", X_train.shape, X_val.shape, X_test.shape)
```

### Optional correlation screening (lightweight dimensionality reduction)

```python
# Use TRAIN only to avoid leakage
corr = X_train.corr(numeric_only=True).abs()

# Drop highly correlated features (keeps model compact + defensible)
TH = 0.98
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [c for c in upper.columns if any(upper[c] > TH)]

X_train_f = X_train.drop(columns=to_drop)
X_val_f   = X_val.drop(columns=to_drop)
X_test_f  = X_test.drop(columns=to_drop)

print("Dropped (high corr):", len(to_drop), "| Remaining features:", X_train_f.shape[1])
```

### Train multi-output model (robust + simple)

```python
# Pipeline: median impute -> strong tabular regressor
reg = HistGradientBoostingRegressor(
    max_depth=6,
    learning_rate=0.05,
    max_iter=400,
    random_state=42
)

model = MultiOutputRegressor(
    Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("reg", reg)
    ]),
    n_jobs=-1
)

model.fit(X_train_f, Y_train)
print("Model trained.")
```

### Results table (MAE/RMSE) + grouped summary

```python
Yhat_test = pd.DataFrame(model.predict(X_test_f), index=Y_test.index, columns=Y_test.columns)

def per_target_metrics(Y_true, Y_pred):
    mae  = mean_absolute_error(Y_true, Y_pred, multioutput="raw_values")
    rmse = np.sqrt(mean_squared_error(Y_true, Y_pred, multioutput="raw_values"))
    out = pd.DataFrame({"Target": Y_true.columns, "MAE": mae, "RMSE": rmse})
    return out.sort_values("RMSE", ascending=False)

metrics = per_target_metrics(Y_test, Yhat_test)

# Grouping for Table I
def group_name(col):
    if col.startswith("UTHD"):
        return "Voltage THD (UTHD)"
    if "IHD3" in col or "IHD9" in col or "IHD15" in col:
        return "Triplen IHD (3/9/15)"
    if "IHD5" in col or "IHD7" in col or "IHD11" in col:
        return "Odd IHD (5/7/11)"
    return "Other"

metrics["Group"] = metrics["Target"].apply(group_name)

summary = metrics.groupby("Group")[["MAE","RMSE"]].agg(["mean","max"]).round(4)
print("=== Grouped summary (Table I style) ===")
display(summary)

# Full metrics table (use this to paste into paper appendix or extra table)
display(metrics.head(10))
```

```python
# ============================================================
# WINDOW SUMMARY: % Normal / Watch / Exceed inside chosen window
# ============================================================

def state_percentages(series, watch_lo, exceed_hi):
    """
    Returns counts and percentages for Normal / Watch / Exceed.
    Normal: < watch_lo
    Watch : [watch_lo, exceed_hi)
    Exceed: >= exceed_hi
    """
    s = pd.Series(series).dropna().astype(float)
    n = len(s)
    if n == 0:
        return {"n": 0, "normal_%": np.nan, "watch_%": np.nan, "exceed_%": np.nan}

    normal = (s < watch_lo).sum()
    watch  = ((s >= watch_lo) & (s < exceed_hi)).sum()
    exceed = (s >= exceed_hi).sum()

    return {
        "n": n,
        "normal_%": 100.0 * normal / n,
        "watch_%":  100.0 * watch  / n,
        "exceed_%": 100.0 * exceed / n
    }

def print_window_summary(name, idx, series_dict, thresholds_dict):
    """
    series_dict: {label: series_aligned_to_full_index}
    thresholds_dict: {label: (watch_lo, exceed_hi)}
    """
    if idx is None or len(idx) == 0:
        print(f"[{name}] No index window available.")
        return

    start, end = idx.min(), idx.max()
    duration_min = (end - start).total_seconds() / 60.0 if hasattr(end - start, "total_seconds") else np.nan

    print(f"\n=== {name} WINDOW SUMMARY ===")
    print(f"Window: {start} → {end}  |  ~{duration_min:.1f} min  |  samples={len(idx)}")

    rows = []
    for label, s in series_dict.items():
        if s is None:
            continue
        if label not in thresholds_dict:
            continue
        wl, eh = thresholds_dict[label]
        stats = state_percentages(s.loc[idx] if hasattr(s, "loc") else pd.Series(s, index=idx), wl, eh)
        rows.append([label, stats["n"], stats["normal_%"], stats["watch_%"], stats["exceed_%"]])

    if rows:
        out = pd.DataFrame(rows, columns=["Signal", "N", "Normal %", "Watch %", "Exceed %"])
        display(out.round(2))
    else:
        print("No matching series/threshold pairs found for this window.")


# -------------------------------
# NOW window summary (if available)
# -------------------------------
if "IDX_NOW" in globals():
    now_series = {
        "Triplen IHD (Now)": trip_now,
        "Odd IHD (Now)": odd_now,
        "Voltage THD max (Now)": uthd_now
    }
    now_thresholds = {
        "Triplen IHD (Now)": (IHD_WATCH_LO, IHD_EXCEED),
        "Odd IHD (Now)": (IHD_WATCH_LO, IHD_EXCEED),
        "Voltage THD max (Now)": (THD_WATCH_LO, THD_EXCEED)
    }
    print_window_summary("NOW", IDX_NOW, now_series, now_thresholds)


# --------------------------------
# TEST window summary (if available)
# --------------------------------
if "IDX_TEST" in globals():
    test_series = {
        "Voltage THD max (Now)": uthd_now_test,
        "Voltage THD max (True t+60s)": uthd_true_max_future,
        "Voltage THD max (Pred t+60s)": uthd_pred_max_future,

        "Triplen IHD (Now)": trip_now_test,
        "Triplen IHD (True t+60s)": trip_true_future,
        "Triplen IHD (Pred t+60s)": trip_pred_future,

        "Odd IHD (Now)": odd_now_test,
        "Odd IHD (True t+60s)": odd_true_future,
        "Odd IHD (Pred t+60s)": odd_pred_future,
    }
    test_thresholds = {
        "Voltage THD max (Now)": (THD_WATCH_LO, THD_EXCEED),
        "Voltage THD max (True t+60s)": (THD_WATCH_LO, THD_EXCEED),
        "Voltage THD max (Pred t+60s)": (THD_WATCH_LO, THD_EXCEED),

        "Triplen IHD (Now)": (IHD_WATCH_LO, IHD_EXCEED),
        "Triplen IHD (True t+60s)": (IHD_WATCH_LO, IHD_EXCEED),
        "Triplen IHD (Pred t+60s)": (IHD_WATCH_LO, IHD_EXCEED),

        "Odd IHD (Now)": (IHD_WATCH_LO, IHD_EXCEED),
        "Odd IHD (True t+60s)": (IHD_WATCH_LO, IHD_EXCEED),
        "Odd IHD (Pred t+60s)": (IHD_WATCH_LO, IHD_EXCEED),
    }
    print_window_summary("TEST", IDX_TEST, test_series, test_thresholds)
```
